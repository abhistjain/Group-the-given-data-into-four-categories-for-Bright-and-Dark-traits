```python
df = df.drop(['LOGIC','CONCENTRATION','FOCUS','RETENTION POWER', 'HARD WORKING', 'STUDY HABIT', 'CONSCIOUSNESS', 'SILLY MISTAKES', 'KNOWLEDGE GAP', 'IMPULSIVE', 'LEARNING GAP','SILLY MISTAKES', 'KNOWLEDGE GAP', 'IMPULSIVE','LEARNING GAP' ], axis = 1)
df
```

```python
from sklearn.cluster import KMeans
```

```python
x = df.iloc[:,2:].values
```

```python
km = KMeans(n_clusters = 4)
y = km.fit_predict(x)
```

```python
y
```

```python
x
```

```python
x[y == 3]    # values that comes under the cluster 3
```

```python
plt.scatter(x[y == 0,0],x[y == 0,1], color ='red')
plt.scatter(x[y == 1,0],x[y == 1,1], color ='green')
plt.scatter(x[y == 2,0],x[y == 2,1], color ='blue')
plt.scatter(x[y == 3,0],x[y == 3,1], color = 'black')
```
